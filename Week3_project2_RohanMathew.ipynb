# 2. Most ordered item in choice description

```python
filtered_choice_description = dataset1[dataset1['choice_description'] != 'None Available']
most_ordered_choice_description = filtered_choice_description['choice_description'].value_counts().idxmax()
```

```python
most_ordered_choice_description
```

# 3. Total items ordered

```python
total_items_ordered = dataset1['quantity'].sum()
total_items_ordered
```

# 4. Revenue for the period

```python
total_revenue = (dataset1['quantity'] * dataset1['item_price']).sum()
total_revenue
```

# 5. Total number of orders

```python
total_orders = dataset1['order_id'].nunique()
total_orders
```

# 6. Average revenue amount per order

```python
average_revenue = total_revenue / total_orders
average_revenue
```
